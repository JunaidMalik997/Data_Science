# Distribution Plots

```python
#we will discuss different plot types with seaborn that allows us to visualize the distribution of a dataset
```

```python
import seaborn as sns
```

```python
%matplotlib inline    
#to see our visualizations inside of the notebook
```

```python
#now we will get some data to plot. Seaborn actually comes with a few built in datasets that we can directly load
tips=sns.load_dataset('tips')     #tips is a built in dataset(dataframe)
```

```python
tips.head()
```

```python
#first plot type which is dist plot
#dist plot allows us to show the distribution of a uni-variant(one variable) set of observations

sns.distplot(tips['total_bill'])   #pass in 1 column of dataframe
#we basically get a histogram and a KDE(Kernel Density Estimation) line
```

```python
#to remove kde line
sns.distplot(tips['total_bill'],kde=False)
#so now we get an histogram. And Histogram is just the distibution of where our total bill lies
```

```python
#we can also change the number of bins
sns.distplot(tips['total_bill'],kde=False,bins=30)
#we get a little more information
#most of the bills lie between 10 and 20
```

```python
#joint plot. This plot from seaborn allows us to basically matchup 2 dist plots by variat data(we can combine
#two dist plots together)
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#above we get tip on y axis and total bill along the x axis and in between we have a scatter plot. And this scatter plot
#actually makes sense because it looks like it has a trend that as we go higher in total bills the tips also increase proportionately
```

```python
#joint plot alos gives us additional argument called kind. This kind parameter allows us to effect what is going inside this
#joint plot--> by default it is scattered
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
#hex allows us to create a hexagon distribution representation. Similar to scatter plot, but if it has certain number of 
#points in it, hexagon gets darker otherwise gets lighter
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
#looks like a scatter plot but seaborn will draw a regression line on it
#basically a linear fit to a scatter point data
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
#that allows us to have that 2 dimensional kde, which esentially just shows us the density of where these points match 
#up the most 
#we will using default scatter plot the most (easier to read)
```

```python
#pairplot: It will plot pairwise relationship across the entire dataframe (at least for the numerical columns)
#what pairplot is esentially gonna do the joint plot for every single possible combination of the numerical columns in the
#dataframe
sns.pairplot(tips)
#we just have to pass the dataframe after calling sns.pairplot()
#we get the pair plot for all the numerical column values
```

```python
#we can also pass hue argument to our pairplot. Hue argument is where we would pass in the column name of a categorical column
#Catgorical column means not numerical or continuous but actual categories. For example, the sex column is categorical because
#there are 2 categories in it male and female
sns.pairplot(tips,hue='sex')
#and it will cover the datapoints based on the column we put in hue
```

```python
#Rug Plots: They are actually a very simple concept, but we will use the rug plots to understand the kde plots we saw earlier
sns.rugplot(tips['total_bill'])    #just like dist plot we will pass in the single column
#so what rug plot does is it just draws the dash mark for every point of the column(single variable) value
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
#so the difference between the above histogram(distplot) and the rugplot is that the histogram has the bins and it counts 
#how many dashes(in rug plot) are in the bin and then shows that up as the number in the histogram
```

```python
#so now we would build off of this idea of rug plot to explain what this kde plot is (kernel density estimation)
```

```python
#if we only want kde plot in dist plot
sns.kdeplot(tips['total_bill'])
```

# Categorical Plots

```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips
```

```python
#for categorical plots, we are mainly gonna be concerned about seeing the distribution of a categorical column such as
#sex,smoker,dat, time etc and reference them either to one of the numerical columns or another categorical column.
```

```python
#start off with the most basic categorical plot which is the bar plot
#bar plot is just a general plot that allows us to aggregate the categorical data besed off of some function and by
#default that is the mean. We can think of it as the visualization of the group by action
sns.barplot(x='sex',y='total_bill',data=tips)
#by default it is showing the average/mean of the total bill per categorical column value
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
#this will tell the standard deviation of the total bill column per category 
#by default the estimator function is mean --> above cell (we can even create our own estimator function)
#x is categorical column, y is numerical data column
```

```python
#countplot: In seaborn it is esentially the same as barplot except the estimator is explicitly counting the number of
#occurences and that is why we only set the x value in it and the data
sns.countplot(x='sex',data=tips)
#same as pandas count() function to count from column
```

```python
#box plot and violin plots: These types of plots are used to show the distribution of categorical data. A boxplot is also
#sometimes known as box or whisker plot and it shows the distribution of quantitative data in a way that hopefully
#facilitates the comparison between the variables
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
#these boxplots shows us the quartiles of these datasets while the whiskers extend to show us the rest of the distribution
#except of the points that are outside of these whiskers, called as outliers
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
#hue gives us thw whole another layer of data very quickly
```

```python
#violin plot: It plays a very similar role to a boxplot. 
sns.violinplot(x='day',y='total_bill',data=tips)
#violin plot unlike the boxplot allows us to actually plot all the components that correspond to actual datapoints
#and it is esentially showing the kernel density estimation of the underlying distribution
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
#split=True allows us to have one half be one hue and another half be another. Gives much more information
```

```python
#strip plot: It is going to draw a scatter plot where one variable is categorical
sns.stripplot(x='day',y='total_bill',data=tips)
#scatter plot based of the category.
#jitter argument(study)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
#The `split` parameter has been renamed to `dodge`
```

```python
#some people like to combine the idea of stripplot and violinplot to create something that is known as swarm plot
#swarm plot: Very similar to strip plot except the points are adjusted so that they don't overlap
sns.swarmplot(x='day',y='total_bill',data=tips)
#combination of strip plot and a violin plot
#not advised for very large datasets
```

```python
#we can combine swarm plot with violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
#both pieces of information stacked on top of each other
```

```python
#Factor Plot: The general form of all the plots
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
#so all the plots we have been looking before the most general method is to call sns.catplot and then to specify by the
#kind. For instance, we need to create the violin plot we can set kind='violin'
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```

# Matrix Plots

```python
#we will learn to create matrix plots primarily heat maps using seaborn.
```

```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
#In order for a heat map to work properly, our data should already be in matrix form. Matrix form means that the index 
#name and the column name match up so that the cell value actually indicates something that is relevant to both of those
#names. For example, if we take a look at tips dataframe we have total_bill column which is some sort of label or a variable
#however the row(index) is not an actual variable. In order to get this into matrix form we need to have the variables
#on both the columns and the rows. We can do that by pivot table or trying to get the correlation data
```

```python
tc=tips.corr()
#we have matrix form of data here. What it means, that we have columns and the index variable names actually indicates a
#relevancy to both the column and the index name for the cell where they match up.
```

```python
tc
```

```python
#this is the matrix form we are gonna need our data in if we want to call heat map on it
sns.heatmap(tc)
#all what heatmap does is, it colors in those values based on some sort of gradients
```

```python
#we can pass annotation argument
sns.heatmap(tc,annot=True)
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')    #color map argument
```

```python
flights
```

```python
flights.corr()
```

```python
#we need to get flights data into matrix form into a more presentable form. Meaning we need to convert one of the variables
#in the columns to be index/row name. So we can use pivot table
fp=flights.pivot_table(index='month',columns='year',values='passengers')
#now we have data in some sort of matrix form
```

```python
fp
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp,cmap='magma')
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidth=1)
```

```python
sns.heatmap(fp,cmap='magma',annot=True,annot_kws={"size": 5},linecolor='white',linewidth=1)
```

```python
sns.heatmap(fp)
```

```python
#now let's talk about the second matrix type plot in seaborn which is the cluster map. The cluster map is going to use 
#hierarchical clustering to produce a clustered version of this heatmap.
```

```python
sns.clustermap(fp)
#it basically clusters the information to show columns and rows that are basically similar to each other 
```

```python
sns.clustermap(fp,cmap='coolwarm')
```

```python
#normalize the scale
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
#cluster map is not in order while heat map. Cluster map will order so that we may infer more information and similar group
#are close to each other
```

# Grids

```python
#In seaborn we will learn about grids capability to automate subplots based off of features in our data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
#we are actually going to customize this sort of pairplot, with a lot more control over the actual plot
#we can do that by using pairgrid 
g=sns.PairGrid(iris)      #gives us empty pair grids. Takes all the numerical columns and grids them up (subplots)

g.map(plt.scatter)
#we can map plot types
#it will map scatter plot on everything, diagonal, upper half and lower half diagonal
```

```python
#we can actually specify on what we want on diagonals, upper half and lower half
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
#we have a PairGrid and now we have much more control over the plots, instead of just being scatter plots and histograms
#now we can choose what plot types we want to map
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
#we have an empty grid and we can actually begin to map plot types to this grid 

g.map(sns.distplot,'total_bill')
#we are mapping a plot type i.e. distplot and separating the results based off of by columns and rows. They are labelled for
#us as follows. So we have 3 variables at play here. Rows and columns i.e. what variables we want to separate by for our
#facet grid and when we are mapping we want to add an argument for which we want to map for display.
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')    #this plot type needs more than one argument
#now we have tip vs total_bill and we have separated both on row and column on smoker and time
```

# Regression Plots

```python
#we will discuss regression plots in detail in machine learning section
#so now we will only cover the LM plot function. LM plot allows us to display linear models with seaborn. We will also learn
#how to size and change the aspect ratio of seaborn plots.
```

```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)     #lmplot stands for linear model plot
#we basically have a scatter plot with a linear fit on top of it.
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
#2 scatter plots and 2 linear fits on top of this. As sex has 2 categories male and female.
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
#we can also add matplotlib style markers
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
#markers are a lot bigger now
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
#instead of specifying hue we can start calling factor or grid plot by calling col(column)
#now we have 2 separate column plots separated by sex category
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
#essentially doig facet grid calls.
```

```python
#we can also add hue in lm plot even after adding row and col
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex')
#gives a lot more information
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,height=8)
#we can adjust the aspect and sie/height of our plot
```
